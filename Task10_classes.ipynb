```python
# Base Class: SoulMachine
class SoulMachine:
    def __init__(self, name, purpose):
        self.name = name
        self.purpose = purpose

    def display_info(self):
        return f"Soul Machine: {self.name}, Purpose: {self.purpose}"


# Subclass 1: Emotional Intelligence
class EmotionalIntelligence(SoulMachine):
    def __init__(self, name, purpose, emotion_capability):
        super().__init__(name, purpose)
        self.emotion_capability = emotion_capability

    def display_info(self):
        return f"{super().display_info()}, Emotional Capability: {self.emotion_capability}"


# Subclass 2: Communication Skills
class CommunicationSkills(SoulMachine):
    def __init__(self, name, purpose, language_capability):
        super().__init__(name, purpose)
        self.language_capability = language_capability

    def display_info(self):
        return f"{super().display_info()}, Languages Supported: {self.language_capability}"


# Subclass 3: Learning and Adaptation
class LearningAndAdaptation(SoulMachine):
    def __init__(self, name, purpose, learning_type):
        super().__init__(name, purpose)
        self.learning_type = learning_type

    def display_info(self):
        return f"{super().display_info()}, Learning Type: {self.learning_type}"


# Subclass 4: Physical Interaction
class PhysicalInteraction(SoulMachine):
    def __init__(self, name, purpose, interaction_mode):
        super().__init__(name, purpose)
        self.interaction_mode = interaction_mode

    def display_info(self):
        return f"{super().display_info()}, Interaction Mode: {self.interaction_mode}"


# Subclass 5: Ethical AI
class EthicalAI(SoulMachine):
    def __init__(self, name, purpose, ethical_guidelines):
        super().__init__(name, purpose)
        self.ethical_guidelines = ethical_guidelines

    def display_info(self):
        return f"{super().display_info()}, Ethical Guidelines: {self.ethical_guidelines}"


# Example Usage
# Create instances of Soul Machines
emotional_ai = EmotionalIntelligence("SoulX", "Emotional Support", "Recognizes and responds to  emotions and say hay to Kumail sir.")
language_ai = CommunicationSkills("none", "Customer Support", ["English", "Urdu"])
adaptive_ai = LearningAndAdaptation("SoulX", "Education", "Self-learning through user interaction")
physical_robot = PhysicalInteraction("none", "Healthcare Assistance", "Physical gestures and touch-based interaction")
ethical_ai = EthicalAI("MoralSoul", "soulx", "Adheres to AI safety guidelines and ethical considerations")

# Display details of each Soul Machine
print(emotional_ai.display_info())
print(language_ai.display_info())
print(adaptive_ai.display_info())
print(physical_robot.display_info())
print(ethical_ai.display_info())
```

```python

#(a)
#Base class 1: Employee
class Employee:
    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id

    def display_info(self):
        return f"Employee Name: {self.name}, Employee ID: {self.emp_id}"


# Subclass of Employee: Manager
class Manager(Employee):
    def __init__(self, name, emp_id, department):
        super().__init__(name, emp_id)
        self.department = department

    def display_info(self):
        return f"{super().display_info()}, Department: {self.department}"


# Subclass of Employee: Developer
class Developer(Employee):
    def __init__(self, name, emp_id, programming_language):
        super().__init__(name, emp_id)
        self.programming_language = programming_language

    def display_info(self):
        return f"{super().display_info()}, Programming Language: {self.programming_language}"

# Subclass of Employee: Director
class Director(Employee):
    def __init__(self, name, emp_id, department):
        super().__init__(name, emp_id)
        self.department = department

    def display_info(self):
        return f"{super().display_info()}, Department: {self.department}"

#Example usage
# Employees
manager = Manager("none", "none", "HR")
developer = Developer("none", "none", "Python")
director = Director ("none","none","CEO")
# Displaying information
print(manager.display_info())
print(developer.display_info())
print(director. display_info())
```

```python
#(b)
# Base class 2: Project
class Project:
    def __init__(self, project_name, budget):
        self.project_name = project_name
        self.budget = budget

    def display_info(self):
        return f"Project Name: {self.project_name}, Budget: €{self.budget}"


# Subclass of Project: Software Project
class SoftwareProject(Project):
    def __init__(self, project_name, budget, platform):
        super().__init__(project_name, budget)
        self.platform = platform

    def display_info(self):
        return f"{super().display_info()}, Platform: {self.platform}"


# Subclass of Project: Construction Project
class ConstructionProject(Project):
    def __init__(self, project_name, budget, location):
        super().__init__(project_name, budget)
        self.location = location

    def display_info(self):
        return f"{super().display_info()}, Location: {self.location}"


# Example usage

# Projects
software_project = SoftwareProject("none", 690, "none")
construction_project = ConstructionProject("none", 6900, "none")

# Display information.
print(software_project.display_info())
print(construction_project.display_info())
```

```python

#Base class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        return f"Name: {self.name}, Age: {self.age}"


# Example usage
person = Person("none", 00)
# Base class 1: Nervous System
class NervousSystem:
    def __init__(self, function):
        self.function = function

    def display_info(self):
        return f"Nervous System Function: {self.function}"


# Subclass of Nervous System
class Brain(NervousSystem):
    def __init__(self, function, region):
        super().__init__(function)
        self.region = region

    def display_info(self):
        return f"{super().display_info()}, Brain Region: {self.region}"


class SpinalCord(NervousSystem):
    def __init__(self, function, length):
        super().__init__(function)
        self.length = length

    def display_info(self):
        return f"{super().display_info()}, Spinal Cord Length: {self.length} cm"


class Nerves(NervousSystem):
    def __init__(self, function, type):
        super().__init__(function)
        self.type = type

    def display_info(self):
        return f"{super().display_info()}, Nerve Type: {self.type}"


# Base class 2: Circulatory System
class CirculatorySystem:
    def __init__(self, function):
        self.function = function

    def display_info(self):
        return f"Circulatory System Function: {self.function}"


# Subclass of Circulatory System
class Heart(CirculatorySystem):
    def __init__(self, function, chambers):
        super().__init__(function)
        self.chambers = chambers

    def display_info(self):
        return f"{super().display_info()}, Heart Chambers: {self.chambers}"


class Blood(CirculatorySystem):
    def __init__(self, function, type):
        super().__init__(function)
        self.type = type

    def display_info(self):
        return f"{super().display_info()}, Blood Type: {self.type}"


class BloodVessels(CirculatorySystem):
    def __init__(self, function, length):
        super().__init__(function)
        self.length = length

    def display_info(self):
        return f"{super().display_info()}, Blood Vessel Length: {self.length} km"


# Base class 3: Digestive System
class DigestiveSystem:
    def __init__(self, function):
        self.function = function

    def display_info(self):
        return f"Digestive System Function: {self.function}"


# Subclass of Digestive System
class Stomach(DigestiveSystem):
    def __init__(self, function, pH_level):
        super().__init__(function)
        self.pH_level = pH_level

    def display_info(self):
        return f"{super().display_info()}, Stomach pH Level: {self.pH_level}"


class Intestines(DigestiveSystem):
    def __init__(self, function, length):
        super().__init__(function)
        self.length = length

    def display_info(self):
        return f"{super().display_info()}, Intestine Length: {self.length} meters"


class Liver(DigestiveSystem):
    def __init__(self, function, bile_production):
        super().__init__(function)
        self.bile_production = bile_production

    def display_info(self):
        return f"{super().display_info()}, Bile Production: {self.bile_production} mL/day"


# Example usage
# Nervous System
brain = Brain("Controls body functions", "Cerebrum")
spinal_cord = SpinalCord("Transmits signals", 0)
nerves = Nerves("Transmits messages", "Sensory")

# Circulatory System
heart = Heart("Pumps blood", 0)
blood = Blood("Transports oxygen", "none")
blood_vessels = BloodVessels("Circulates blood", 0)

# Digestive System
stomach = Stomach("Digests food", 0)
intestines = Intestines("Absorbs nutrients", 0)
liver = Liver("Produces bile", 0)

# Displaying information
print(person.display_info())
print(brain.display_info())
print(spinal_cord.display_info())
print(nerves.display_info())

print(heart.display_info())
print(blood.display_info())
print(blood_vessels.display_info())

print(stomach.display_info())
print(intestines.display_info())
print(liver.display_info())
```

```python
# Base class 1: Planet
class Planet:
    def __init__(self, name, distance_from_sun):
        self.name = name
        self.distance_from_sun = distance_from_sun  # Distance in million km

    def display_info(self):
        return f"Planet: {self.name}, Distance from Sun: {self.distance_from_sun} million km"


# Subclass of Planet
class TerrestrialPlanet(Planet):
    def __init__(self, name, distance_from_sun, has_atmosphere):
        super().__init__(name, distance_from_sun)
        self.has_atmosphere = has_atmosphere

    def display_info(self):
        return f"{super().display_info()}, Has Atmosphere: {self.has_atmosphere}"


class GasGiant(Planet):
    def __init__(self, name, distance_from_sun, ring_system):
        super().__init__(name, distance_from_sun)
        self.ring_system = ring_system

    def display_info(self):
        return f"{super().display_info()}, Ring System: {self.ring_system}"


class DwarfPlanet(Planet):
    def __init__(self, name, distance_from_sun, size):
        super().__init__(name, distance_from_sun)
        self.size = size  # Size in km diameter

    def display_info(self):
        return f"{super().display_info()}, Size: {self.size} km"


# Base class 2: Star
class Star:
    def __init__(self, name, type, temperature):
        self.name = name
        self.type = type  # Type of star (e.g., main-sequence, red giant)
        self.temperature = temperature  # Temperature in Kelvin

    def display_info(self):
        return f"Star: {self.name}, Type: {self.type}, Temperature: {self.temperature} K"


# Subclass of Star
class MainSequenceStar(Star):
    def __init__(self, name, temperature, lifespan):
        super().__init__(name, "Main-Sequence", temperature)
        self.lifespan = lifespan  # Lifespan in billion years

    def display_info(self):
        return f"{super().display_info()}, Lifespan: {self.lifespan} billion years"


class RedGiantStar(Star):
    def __init__(self, name, temperature, size):
        super().__init__(name, "Red Giant", temperature)
        self.size = size  # Size in solar radii

    def display_info(self):
        return f"{super().display_info()}, Size: {self.size} solar radii"


class WhiteDwarfStar(Star):
    def __init__(self, name, temperature, mass):
        super().__init__(name, "White Dwarf", temperature)
        self.mass = mass  # Mass in solar masses

    def display_info(self):
        return f"{super().display_info()}, Mass: {self.mass} solar masses"


# Base class 3: Moon
class Moon:
    def __init__(self, name, planet_orbits, diameter):
        self.name = name
        self.planet_orbits = planet_orbits  # Name of planet it orbits
        self.diameter = diameter  # Diameter in km

    def display_info(self):
        return f"Moon: {self.name}, Orbits: {self.planet_orbits}, Diameter: {self.diameter} km"


# Subclass of Moon
class RockyMoon(Moon):
    def __init__(self, name, planet_orbits, diameter, terrain_type):
        super().__init__(name, planet_orbits, diameter)
        self.terrain_type = terrain_type

    def display_info(self):
        return f"{super().display_info()}, Terrain Type: {self.terrain_type}"


class IcyMoon(Moon):
    def __init__(self, name, planet_orbits, diameter, water_content):
        super().__init__(name, planet_orbits, diameter)
        self.water_content = water_content  # Water content in %

    def display_info(self):
        return f"{super().display_info()}, Water Content: {self.water_content}%"


class VolcanicMoon(Moon):
    def __init__(self, name, planet_orbits, diameter, active_volcanoes):
        super().__init__(name, planet_orbits, diameter)
        self.active_volcanoes = active_volcanoes  # Number of active volcanoes

    def display_info(self):
        return f"{super().display_info()}, Active Volcanoes: {self.active_volcanoes}"


# Example usage
# Planets
earth = TerrestrialPlanet("Earth", 149.6, True)
jupiter = GasGiant("Jupiter", 778.5, True)
pluto = DwarfPlanet("Pluto", 5906.4, 2376)

# Stars
sun = MainSequenceStar("Sun", 5778, 10)
betelgeuse = RedGiantStar("Betelgeuse", 3500, 950)
sirius_b = WhiteDwarfStar("Sirius B", 25000, 1.02)

# Moons
moon = RockyMoon("Moon", "Earth", 3474.2, "Rocky")
europa = IcyMoon("Europa", "Jupiter", 3121.6, 20)
io = VolcanicMoon("Io", "Jupiter", 3643.2, 400)

# Displaying information
print(earth.display_info())
print(jupiter.display_info())
print(pluto.display_info())

print(sun.display_info())
print(betelgeuse.display_info())
print(sirius_b.display_info())

print(moon.display_info())
print(europa.display_info())
print(io.display_info())
```
